```python
# Q.2.Implement K-Nearest Neighbor classifier on the iris dataset and analyze the performance
# using accuracy, while varying the number of neighbors (e.g. 1 – 250). Also print the
# neighbor(s) having the highest accuracy and those with the lowest accuracy.

import pandas as pd
df = pd.read_csv(r'iris.data')
names_of_coloumns = ['sepal_length','sepal_width','petal_length','petal_width','class']
df.columns = names_of_coloumns
print(df)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import numpy as np

features = df.iloc[:, 0:4].values
labels = df.iloc[:, 4].values

X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.2, random_state=2)

accuracies = []
for neighbors in range(1, min(251, len(X_train) + 1)):
    classifier = KNeighborsClassifier(n_neighbors=neighbors)
    classifier.fit(X_train, y_train)
    y_pred_train = classifier.predict(X_train)
    accuracy = accuracy_score(y_train, y_pred_train)
    accuracies.append(accuracy)

plt.figure(figsize=(12, 8))
plt.plot(range(1, min(251, len(X_train) + 1)), accuracies, marker='o', linestyle='--', color='green')
plt.title('Accuracy vs. Number of Neighbors')
plt.xlabel('Number of Neighbors')
plt.ylabel('Accuracy')
plt.grid(True)
plt.show()
```

```python
max=accuracies[0]
for j in accuracies:
  if(j>max):
    max=j
    k=k+1
print("Neighbour {} highest accuracy {}".format(k,max))
```

```python
for random_seed in range(1, 11):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=random_seed)
    knn_classifier = KNeighborsClassifier(n_neighbors=2)
    knn_classifier.fit(X_train, y_train)
    predictions = knn_classifier.predict(X_test)
    print("Accuracy Score (Seed {}): {}".format(random_seed, accuracy_score(y_test, predictions)))
```

```python
from sklearn.metrics import classification_report, confusion_matrix

for random_seed in range(1, 11):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=random_seed)
    knn_classifier = KNeighborsClassifier(n_neighbors=2)
    knn_classifier.fit(X_train, y_train)
    predictions = knn_classifier.predict(X_test)

    print("Classification Report (Seed {}):".format(random_seed))
    print(classification_report(y_test, predictions))

    print("Confusion Matrix (Seed {}):".format(random_seed))
    print(confusion_matrix(y_test, predictions))
```
